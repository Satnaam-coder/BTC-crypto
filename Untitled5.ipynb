```python
!pip install yfinance pandas numpy matplotlib scikit-learn statsmodels prophet tensorflow

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
btc = yf.download('BTC-USD', start='2015-01-01', end=pd.Timestamp.today().strftime('%Y-%m-%d'))
btc = btc.reset_index()
btc = btc[['Date', 'Close', 'Volume']]
btc.columns = ['dt', 'close', 'vol']
btc['dt'] = pd.to_datetime(btc['dt'])
btc.set_index('dt', inplace=True)

print("Missing values:", btc.isnull().sum())
btc = btc.bfill()
btc = btc.ffill()
print("Missing after fill:", btc.isnull().sum())
```

```python
btc['close'].plot(title='BTC Close Price', figsize=(10,4))
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()

btc['vol'].plot(title='BTC Volume', figsize=(10,4))
plt.xlabel('Year')
plt.ylabel('Volume')
plt.show()
```

```python
data = btc['close']

n = len(data)
train = data[:int(0.8*n)]
test = data[int(0.8*n):]
```

```python
from statsmodels.tsa.arima.model import ARIMA

arima = ARIMA(train, order=(5,1,0))
arima_fit = arima.fit()
arima_pred = arima_fit.forecast(len(test))
mae_arima = mean_absolute_error(test, arima_pred)
print("ARIMA MAE:", mae_arima)

plt.plot(test.index, test, label='Actual')
plt.plot(test.index, arima_pred, label='ARIMA Predict')
plt.legend()
plt.title('ARIMA: Actual vs Predicted')
plt.show()
```

```python
dfp = data.reset_index()
dfp.columns = ['ds','y']

train_p = dfp.iloc[:int(0.8*n)]
test_p = dfp.iloc[int(0.8*n):]

m = Prophet()
m.fit(train_p)

future = m.make_future_dataframe(periods=len(test_p))
fcast = m.predict(future)

plt.plot(test_p['ds'], test_p['y'], label='Actual')
plt.plot(test_p['ds'], fcast['yhat'].iloc[-len(test_p):], label='Prophet Predict')
plt.legend()
plt.title('Prophet: Actual vs Predicted')
plt.show()

mae_prophet = mean_absolute_error(test_p['y'], fcast['yhat'].iloc[-len(test_p):])
print("Prophet MAE:", mae_prophet)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(data.values.reshape(-1,1))

timesteps = 60
X, y = [], []
for i in range(timesteps, len(scaled)):
    X.append(scaled[i-timesteps:i, 0])
    y.append(scaled[i, 0])
X, y = np.array(X), np.array(y)

xt = X[:int(0.8*len(X))]
yt = y[:int(0.8*len(y))]
xtest = X[int(0.8*len(X)):]
ytest = y[int(0.8*len(y)):]

xt = xt.reshape((xt.shape[0], xt.shape[1], 1))
xtest = xtest.reshape((xtest.shape[0], xtest.shape[1], 1))

model = keras.Sequential()
model.add(layers.LSTM(50, return_sequences=True, input_shape=(xt.shape[1], 1)))
model.add(layers.LSTM(50))
model.add(layers.Dense(1))
model.compile(optimizer='adam', loss='mae')
model.fit(xt, yt, epochs=5, batch_size=32, verbose=1)

ypred = model.predict(xtest)
ypred_res = scaler.inverse_transform(ypred)
ytest_res = scaler.inverse_transform(ytest.reshape(-1,1))

plt.plot(ytest_res, label='Actual')
plt.plot(ypred_res, label='LSTM Predict')
plt.legend()
plt.title('LSTM: Actual vs Predicted')
plt.show()

mae_lstm = mean_absolute_error(ytest_res, ypred_res)
print("LSTM MAE:", mae_lstm)
```

```python
print("MAE: ARIMA", mae_arima)
print("MAE: Prophet", mae_prophet)
print("MAE: LSTM", mae_lstm)
```
